```python
#Que No 6
#Write a python program to check whether two lists are circularly identical.
a=[3,4,5,1,2]
b=[1,2,3,4,5]
c=[]
if len(a) == len(b):
  for i in range(len(a)):
    for j in range(len(b)):
      if(a[i]==a[j]):
        c.append(a[i])
  if c==a or c==b:
    print("Lists are circularly identical")
  else:
    print("Lists are not circularly identical")
else:
  print("Lists are not circularly identical")
```

```python
#Que No 9
#Write a Python program to generate the combinations of n distinct objects taken from the elements of a given list. Original
#list: [1, 2, 3, 4, 5, 6, 7, 8, 9] Combinations of 2 distinct objects: [1, 2] [1, 3] [1, 4] [1, 5] .... [7, 8] [7, 9] [8, 9]

ls=[1, 2, 3, 4, 5, 6, 7, 8, 9]
ls1 = []

for i in ls:
  for j in ls:
    ls2 = []
    if (i == j):
     continue
    elif (i != j):
      ls2.append(i)
      ls2.append(j)
    ls1.append(ls2)
print(ls1)
```

```python
#Que No 10
#Write a Python program to remove duplicates from a list of lists.
#Sample list : [[10, 20], [40], [30, 56, 25], [10, 20], [33],[40]]
#New List : [[10, 20], [30, 56, 25], [33], [40]]

ls=[[10, 20], [40], [30, 56, 25], [10, 20], [33],[40]]
ls1=[]
for i in ls:
  if i not in ls1:
    ls1.append(i)
print(ls1)
```
