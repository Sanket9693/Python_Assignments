```python
class Bank:
  acc_no=1
  def __init__(self):
    self.name=input("Enter name: ")
    Bank.acc_no+=100
    self.balance=0
  
  def deposit(self):
    print("Name of account holder:",self.name,"Account No:",Bank.acc_no)
    amount=float(input("Enter amount to deposit: "))
    if(amount<0):
      print("Amount can not be negative")
    elif(amount>=2000):
      self.balance+=amount
      print("Amount deposited:",amount)
    else:
      print("The amount should be at least Rs.2000")
  
  def withdraw(self):
    print("Name of account holder:",self.name,"\nAccount No:",Bank.acc_no,"\nAccount Balace:",self.balance)
    amount=float(input("Enter amount to withdraw: "))
    if(self.balance>=amount):
      self.balance-=amount
      print("Amount withdraw:",amount,"\nRemaining balace:",self.balance)
    else:
      print("Insufficient Balance")

  def display(self):
    print("Account No:",Bank.acc_no,"\nName of account holder:",self.name,"\nTotal Balance:",self.balance)
```

```python
# Write a Rectangle class in Python language, allowing you to build a rectangle with length and width attributes.
# Create a Perimeter() method to calculate the perimeter of the rectangle and a Area() method to calculate the area of ​​the rectangle.
# Create a method display() that display the length, width, perimeter and area of an object created using an instantiation on rectangle class.
# Create a Parallelepipede child class inheriting from the Rectangle class and with a height attribute and 
# another Volume() method to calculate the volume of the Parallelepiped.

class Rectangle:
  def create_Rect(self):
    self.length=int(input("Enter the length: "))
    self.width=int(input("Enter the width: "))
  
  def perimeter(self):
    self.p=2*(self.length+self.width)
  
  def area(self):
    self.a=self.length*self.width

  def display(self):
    print("Length of rectangle:",self.length,"\nWidth of rectangle:",self.width,"\nPerimeter of rectangle=",self.p,"\nArea of rectangle=",self.a)
```

```python
class Parallelepiped(Rectangle):
  def volume(self):
    super().create_Rect()
    self.height=int(input("Enter the height: "))
    self.v=self.length * self.width * self.height
    print("Volume of parellelepiped is",self.v)
```

```python
s1=Rectangle()
s2=Parallelepiped()
```

```python
s1.create_Rect()
s1.perimeter()
s1.area()
s1.display()
```

```python
s2.volume()
```
